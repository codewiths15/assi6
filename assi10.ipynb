```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IRIS.csv')
df.head()
```

```python
df.info()
```

```python
fig,axes = plt.subplots(2,2, figsize=(15,15))
axes[0,0].set_title('Distribution of sepal length')
axes[0,0].hist(df['sepal_length'])
axes[0,1].set_title('Distribution of sepal length')
axes[0,1].hist(df['sepal_width'])
axes[1,0].set_title('Distribution of sepal length')
axes[1,0].hist(df['petal_length'])
axes[1,1].set_title('Distribution of sepal length')
axes[1,1].hist(df['petal_width'])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
axes[0,0].set_title("Distribution of Sepal Length")
sns.boxplot(  y="sepal_length", x= "species", data=df,  orient='v', ax=axes[0,0], hue='species')
axes[0,1].set_title("Distribution of Sepal Length")
sns.boxplot(  y="sepal_width", x= "species", data=df,  orient='v', ax=axes[0,1], hue='species' )
axes[1,0].set_title("Distribution of Sepal Length")
sns.boxplot(  y="petal_length", x= "species", data=df,  orient='v', ax=axes[1,0], hue='species' )
axes[1,1].set_title("Distribution of Sepal Length")
sns.boxplot(  y="petal_width", x= "species", data=df,  orient='v', ax=axes[1,1],hue='species' )
plt.show()
```
