```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Boston.csv')
df1=pd.read_csv('Boston1.csv')
```

```python
X= df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']]
Y=df['medv']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, Y_train)
y_pred=lr.predict(X_test)
```

```python
print(y_pred)
```

```python
sns.set(style="whitegrid")

# Loop through each independent variable and plot against 'medv'
for col in df.columns[1:-1]:
    sns.lmplot(x=col, y='medv', data=df, line_kws={'color': 'red'}, scatter_kws={'alpha':0.5})
    plt.title(f'Linear Regression Plot of {col} vs medv')
    plt.xlabel(col)
    plt.ylabel('medv')
    plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_squared_error, r2_score
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
